```python
import seaborn as sb
import pandas as pd

iris_dataset=sb.load_dataset('iris')
iris_dataset.head()
df=pd.DataFrame(iris_dataset[["sepal_length","sepal_width","petal_length","petal_width","species"]])
iris_target=df["species"]
iris_target.head()
iris_target.duplicated().sum()
iris_dataset.drop_duplicates()
iris_dataset.info()
iris_dataset.describe()
val=iris_dataset["species"].value_counts()
display(val)
```

```python
iris_dataset.head()
x=df.drop(columns=["species"],axis=1)
y=df["species"]
x.head()
y.head()
from sklearn.preprocessing import OneHotEncoder
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
from sklearn.preprocessing import OneHotEncoder

ohe=OneHotEncoder()
# y_train=ohe.fit_transform(y_train.values.reshape(-1,1)).toarray() # Removed OneHotEncoder
# y_test=ohe.transform(y_test.values.reshape(-1,1)).toarray()     # Removed OneHotEncoder
y_train
```

```python
from sklearn.linear_model import LogisticRegression;import numpy as np
model=LogisticRegression(max_iter=200)
model.fit(x_train,y_train)
```

```python
accuracy = model.score(x_test, y_test)
print(f"Accuracy Score: {accuracy}")
```

```python
import joblib
joblib.dump(model,"iris_model.pkl")
```
